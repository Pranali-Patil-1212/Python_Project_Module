# Empolyee Management System



```python
import sqlite3
```

```python
conn  = sqlite3.connect('company.db')
cursor = conn.cursor()
```

```python
conn.execute("""
CREATE TABLE IF NOT EXISTS employees (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    role TEXT NOT NULL,
    salary REAL
)
""")
conn.commit()
```

```python
def add_employee(name, role, salary):
    conn.execute("INSERT INTO employees (name, role, salary) VALUES (?, ?, ?)", (name, role, salary))

    conn.commit()
    print("Employee added successfully!\n")
```

```python
def view_employees():
    slct = conn.execute("SELECT * FROM employees")
    #rows = conn.fetchall()
    print("\n Empolyee Records:")
    for row in slct:
        print(f"ID: {row[0]}, Name: {row[1]}, Role: {row[2]}, Salary: {row[3]}")
    print()





```

```python
def update_employee(id, name, role, salary):
    conn.execute("UPDATE employees SET name=?, role=?, salary=? WHERE id=?", (name, role, salary, id))
    conn.commit()
    print("Employee updated successfully!\n")
```

```python
def delete_employee(id):
    conn.execute("DELETE FROM employees WHERE id=?", (id,))
    conn.commit()
    print("Employee deleted successfully!\n")
```

```python
def menu():
  while True:
    print("Employee Management System")
    print("1. Add Employee")
    print("2. View Employees")
    print("3. Update Employee")
    print("4. Delete Employee")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice=="1":
      name = input("Enter Employee Name: ")
      role = input("Enter Employee Role: ")
      salary = float(input("Enter Employee Salary: "))
      add_employee(name, role, salary)

    elif choice=="2":
      view_employees()

    elif choice=="3":
      id = int(input("Enter Employee ID to update: "))
      name = input("Enter New Employee Name: ")
      role = input("Enter New Employee Role: ")
      salary = float(input("Enter New Employee Salary: "))
      update_employee(id, name, role, salary)

    elif choice=="4":
      id = int(input("Enter Employee ID to delete: "))
      delete_employee(id)

    elif choice=="5":
      print("Exiting the program.")
      break

    else:
      print("Invalid choice. Please try again.")
```

```python
menu()
```

```python
conn.close()
```

# Inventory Management(Add/ Remove Items)
A Shop keeps a list of items in stock

```python
inventory = ["apple", "banana"]
print("Inventory Management System")
while True:
  print("1. Add new item")
  print("2. Remove sold item")
  print("3. Show")
  print("4. Exit")
  choice = input("Enter your choice:")
  if choice == "1":  #Add new items
    new_item = input("Enter new item to add in list")
    inventory.append(new_item)
    print("Item added successfully")
  elif choice == "2":  #Remove sold item
    sold_item = input("enter the name of sold item to remove")
    inventory.remove(sold_item)
    print("Item removed successfully")
  elif choice =="3":
    print(inventory)
  elif choice=="4":
    print("Exiting the program")
    break

  else:
    print("Invalid choice")
```

# Real-Time Chat Application (WebSocket + FastAPI)

```python
from fastapi import FastAPI, WebSocket
import uvicorn

app = FastAPI()

@app.websocket("/ws")
async def chat(websocket: WebSocket):
  await websocket.accept()
  while True:
    data = await websocket.receive_text()
    await websocket.send_text(f"Message received:{data}")

if __name__ == "__main__":
  uvicorn.run("chat:app", host="0.0.0.0", port=8000, reload=True)
```
